# Pocket2Mol — Molecule Generation

This notebook documents the setup and execution of Pocket2Mol for 
de novo molecule generation targeting the A2A adenosine receptor (PDB: 4EIY).

---

> **Mac ARM64 Note:** Pocket2Mol was developed for Linux/CUDA. 
> The following files in `pocket2mol/` are provided for Mac ARM64 compatibility:
> - `environment_mac_arm64.yml` — replaces `env_cuda113.yml`
> - `maskfill_mac_arm64_cpu.py` — replaces `models/maskfill.py`

---

> **Downloads:**
> - Original repo: https://github.com/pengxingang/Pocket2Mol
> - Pretrained checkpoint: https://drive.google.com/drive/folders/1KfdOczjUPITPhIvCuBmnj4xFTV-iI2xB

## 1. Repository Setup

```bash
cd pocket2mol/
git clone https://github.com/pengxingang/Pocket2Mol.git
cd Pocket2Mol
```

**Pretrained Checkpoint**

After downloading (see links above), place the file here:
`pocket2mol/Pocket2Mol/ckpt/pretrained_Pocket2Mol.pt`

> The checkpoint contains learned parameters from training on ~100K protein-ligand complexes.

## 2. Environment Setup

```bash
# Create conda environment
conda env create -f pocket2mol/environment_mac_arm64.yml

# Activate
conda activate Pocket2Mol

## 3. Mac ARM64 Fix

See Mac ARM64 compatibility files listed in the introduction above. 
Copy the provided file into the Pocket2Mol directory — **filename must match exactly**:

```bash
# Replace maskfill.py with CPU-compatible version
# Note: filename must be exactly 'maskfill.py' — Pocket2Mol looks for this exact name
cp ../../maskfill_mac_arm64_cpu.py models/maskfill.py
```
