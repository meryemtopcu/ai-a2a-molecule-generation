# Pocket2Mol — Molecule Generation

This notebook documents the setup and execution of Pocket2Mol for 
de novo molecule generation targeting the A2A adenosine receptor (PDB: 4EIY).

## 1. Environment Setup

The original `env_cuda113.yml` was written for Linux/CUDA and does not 
work on Mac ARM64. A modified environment file was created for Mac:
`pocket2mol/Pocket2Mol/env_cuda113 2.yml`

```bash
conda env create -f "pocket2mol/Pocket2Mol/env_cuda113 2.yml" -n Pocket2Mol
conda activate Pocket2Mol
```

Additional packages not included in the original environment file were 
installed manually and added:
`easydict`, `lmdb`, `torch-geometric==2.3.0`, `torch-scatter`, 
`torch-sparse`, `torch-cluster`, `numpy<2`.(see `pocket2mol/requirements.txt`).


## 2. Repository Setup

```bash
cd pocket2mol/
git clone https://github.com/pengxingang/Pocket2Mol.git
cd Pocket2Mol
```

Checkpoint downloaded from Google Drive and placed in `ckpt/pretrained_Pocket2Mol.pt`.

## 3. Mac ARM64 Fix

Pocket2Mol was originally written for CUDA (NVIDIA GPU). 
Since Mac ARM64 does not support CUDA, all `.to('cuda')` 
calls were replaced with `.to('cpu')`:

```bash
sed -i '' 's/\.to('\''cuda'\'')/\.to('\''cpu'\'')/g' models/maskfill.py
```

## 4. Molecule Generation

```bash
python sample_for_pdb.py \
    --config configs/sample.yml \
    --pdb_path ../../data/4eiy_clean.pdb \
    --center="-0.42,8.53,17.13" \
    --device cpu \
    --outdir ../../outputs/pocket2mol
```

### Parameters (from configs/sample.yml)

| Parameter | Value | Description |
|---|---|---|
| num_samples | 100 | Number of molecules to generate |
| beam_size | 500 | Beam search width |
| max_steps | 50 | Maximum atoms per molecule |
| seed | 2020 | Random seed for reproducibility |
