```python
# Load both CSVs and compare
import pandas as pd

my_df = pd.read_csv('outputs/pocket2mol/pocket2mol_generated.csv')
friend_df = pd.read_csv('results/pocket2mol_generated.csv')

print("My molecules:", len(my_df))
print("Friend's molecules:", len(friend_df))
print("\nMy SMILES (first 5):")
print(my_df['smiles'].head())
print("\nFriend's SMILES (first 5):")
print(friend_df['smiles'].head())
```

```python
# Find common molecules between both runs
my_smiles = set(my_df['smiles'].tolist())
friend_smiles = set(friend_df['smiles'].tolist())

common = my_smiles.intersection(friend_smiles)
print(f"Common molecules: {len(common)}")
print(f"Only in mine: {len(my_smiles - friend_smiles)}")
print(f"Only in friend's: {len(friend_smiles - my_smiles)}")
```

```python
# Check friend's CSV columns and stats
print("Friend's columns:", friend_df.columns.tolist())
print("\nFriend's stats:")
print(friend_df[['MW', 'LogP', 'QED']].describe().round(2))
print("\nMy stats:")
print(my_df[['MW', 'LogP', 'QED']].describe().round(2))
```
